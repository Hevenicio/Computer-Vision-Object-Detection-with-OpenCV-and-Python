### Create Classifier

```python
plateClassifier = cv2.CascadeClassifier('Haarcascades/haarcascade_plate_number.xml')
```

```python
plate2Classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_plate_number.xml')
```

### It's a Kind of Magic

```python
def DetectImage(img):
    plateImg = img.copy()
    
    plateRects = plateClassifier.detectMultiScale(plateImg, 1.1, 1)
    
    for (x, y, w, h) in plateRects:
        cv2.rectangle(plateImg,
                     (x, y),
                     (x + w, y + h),
                     (100, 255, 255),
                     5)
        
    return plateImg
```

```python
def DetectImage2(img2):
    plateImg2 = img2.copy()
    
    plateRects2 = plate2Classifier.detectMultiScale(plateImg2, 1.1, 1)
    
    for (x, y, w, h) in plateRects2:
        cv2.rectangle(plateImg2,
                     (x, y),
                     (x + w, y + h),
                     (0, 255, 25),
                     10)
        
    return plateImg2
```

### Check the Results

```python
result3 = DetectImage(img)

x = Display(result3)

plt.figure(figsize = (10, 8))

plt.imshow(x)
```

```python
result4 = DetectImage2(img2)

y = Display(result4)

plt.figure(figsize = (10, 8))

plt.imshow(y)
```
