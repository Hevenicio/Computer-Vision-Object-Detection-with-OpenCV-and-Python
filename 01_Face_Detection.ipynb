### Import Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
#%matplotlib notebook
%matplotlib inline
```

### Create Classifier

```python
faceClassifier = cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')
```

### Open Image

```python
image = cv2.imread('Images/eye_face.jpg')

fixImg = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(fixImg)
```

### Detect Faces

```python
faces = faceClassifier.detectMultiScale(image, 1.3, 5)
```

### No Faces Detected

```python
if faces is ():
    print('No Faces found')
```

### It's a kind of Magic

```python
def DetectFace(fixImg):
    faceRects = faceClassifier.detectMultiScale(fixImg)
    
    for (x, y, w, h) in faceRects:
        cv2.rectangle(fixImg,
                     (x, y),
                     (x + w, y + h),
                     (255, 0, 0),
                     10)
        
    return fixImg
```

```python
result = DetectFace(fixImg)

plt.imshow(result)
```
