### Import Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
#%matplotlib notebook
%matplotlib inline
```

### Create Classifiers

```python
faceClassifier = cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')
eyeClassifier = cv2.CascadeClassifier('Haarcascades/haarcascade_eye.xml')
```

### Open Image

```python
image = cv2.imread('Images/eye_face.jpg')
```

### Fix Image

```python
fixImage = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

### Face Classifier

```python
faces = faceClassifier.detectMultiScale(fixImage, 1.3, 5)
```

### If No Face Detected

```python
if faces is ():
    print('No Faces found')
```

### It's a Kind of Magic

```python
def DetectFaceEyes(fixImg):
    faceRects = faceClassifier.detectMultiScale(fixImg)
    
    for (x, y, w, h) in faceRects:
        cv2.rectangle(fixImg,
                     (x, y),
                     (x + w, y + h),
                     (255, 0, 0),
                     10)
        
        eyeRects = eyeClassifier.detectMultiScale(fixImg)
    
        for (x, y, w, h) in eyeRects:
            cv2.rectangle(fixImg,
                         (x, y),
                         (x + w, y + h),
                         (0, 255, 0),
                         10)
        
    return fixImg
```

```python
result = DetectFaceEyes(fixImage)

plt.imshow(result)
```
