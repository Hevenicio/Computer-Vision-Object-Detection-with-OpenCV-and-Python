### Import Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
#%matplotlib notebook
%matplotlib inline
```

### Open Image

```python
image = cv2.imread('Images/eye_face.jpg')
```

### Show Image

```python
plt.imshow(image);
```

### Fix Image

```python
fixImage = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

### Show Image

```python
plt.imshow(fixImage);
```

### Create Classifier

```python
eyeClassifier = cv2.CascadeClassifier('Haarcascades/haarcascade_eye.xml')
```

### It's a Kind of Magic

```python
def DetectEyes(fixImg):
    eyeRects = eyeClassifier.detectMultiScale(fixImg)
    
    for (x, y, w, h) in eyeRects:
        cv2.rectangle(fixImg,
                     (x, y),
                     (x + w, y + h),
                     (0, 255, 0),
                     10)
        
    return fixImg
```

### Show Results

```python
result = Detecteyes(fixImage)

plt.imshow(result)
```
